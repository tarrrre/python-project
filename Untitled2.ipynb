```python
import time
import numpy as np
from datetime import datetime
from IPython.display import display, HTML, Audio

# =========================
# ì„¤ì •
# =========================
WORK_MINUTES = 0.5
BREAK_MINUTES = 0.5
CYCLES = 2
ENABLE_SOUND = True


# =========================
# ì†Œë¦¬ ìœ í‹¸
# =========================
def play_wave(wave, sr=44100):
    display(Audio(wave, rate=sr, autoplay=True))

def generate_beep(freq, duration):
    sr = 44100
    t = np.linspace(0, duration, int(sr * duration))
    return np.sin(2 * np.pi * freq * t)

# =========================
# ê¸´ ì•Œë¦¼ìŒ íŒ¨í„´ë“¤
# =========================
def sound_work_end():
    # ë¶€ë“œëŸ¬ìš´ ë²¨ 0.4ì´ˆ Ã— 6íšŒ (ì•½ 2.5ì´ˆ)
    wave = np.concatenate([
        generate_beep(880, 0.4) * np.exp(-2 * np.linspace(0, 0.4, int(44100 * 0.4)))
        for _ in range(6)
    ])
    play_wave(wave)

def sound_break_end():
    # ê°•í•œ ì•ŒëŒ 0.3ì´ˆ Ã— 8íšŒ (ì•½ 2.4ì´ˆ)
    wave = np.concatenate([
        generate_beep(1200, 0.3) * (np.sin(2 * np.pi * 4 * np.linspace(0, 0.3, int(44100 * 0.3))) > 0)
        for _ in range(8)
    ])
    play_wave(wave)

def sound_all_done():
    # ë©œë¡œë”” ë°˜ë³µ (ì•½ 4ì´ˆ)
    sr = 44100
    melody = [(880, 0.3), (1320, 0.3), (1760, 0.4)]
    wave = np.concatenate([
        generate_beep(f, d)
        for _ in range(3)
        for f, d in melody
    ])
    play_wave(wave)

def play_sound(kind):
    if not ENABLE_SOUND:
        return
    if kind == "work_end":
        sound_work_end()
    elif kind == "break_end":
        sound_break_end()
    elif kind == "all_done":
        sound_all_done()

# =========================
# íƒ€ì´ë¨¸
# =========================
def run_timer(minutes, label):
    print(f"\nâ³ {label} ì‹œì‘ ({minutes}ë¶„)")
    time.sleep(minutes * 60)

    now = datetime.now().strftime("%H:%M")
    print(f"ğŸ”” {label} ì¢…ë£Œ - {now}")

    notify(f"{label} ì¢…ë£Œ", f"{now} - ë‹¤ìŒ ë‹¨ê³„ë¡œ ì´ë™í•˜ì„¸ìš”")

    if label == "ì—…ë¬´":
        play_sound("work_end")
    else:
        play_sound("break_end")

# =========================
# í¬ëª¨ë„ë¡œ ì‹¤í–‰
# =========================
def pomodoro(work, rest, cycles):
    print("ğŸ… Colab ì—…ë¬´ ì•Œë¦¼ íƒ€ì´ë¨¸ ì‹œì‘")

    for i in range(1, cycles + 1):
        print(f"\nâ–¶ ì‚¬ì´í´ {i}/{cycles}")
        run_timer(work, "ì—…ë¬´")
        run_timer(rest, "íœ´ì‹")

    print("\nğŸ‰ ëª¨ë“  ì‚¬ì´í´ ì™„ë£Œ")
    notify("íƒ€ì´ë¨¸ ì™„ë£Œ", "ëª¨ë“  ì—…ë¬´ ì‚¬ì´í´ì´ ëë‚¬ìŠµë‹ˆë‹¤ ğŸ‰")
    play_sound("all_done")

# =========================
# ì‹¤í–‰ (âš  í…ŒìŠ¤íŠ¸ëŠ” ì§§ê²Œ)
# =========================
pomodoro(WORK_MINUTES, BREAK_MINUTES, CYCLES)
```
